```python
import pandas as pd
data1 = pd.read_excel('mabse.xlsx')
df1 = pd.DataFrame(data1)
df1.sort_values(by = ['mgtkas_date'])
df1 = df1.loc[df1['mgtkas_date']  >= '2012-01-01']
df1.shape
```

```python
data2 = pd.read_excel('bse500companies.xlsx')
df2 = pd.DataFrame(data2)
df2.head()
```

```python
list1 = df2['CAPS'].tolist()
df3 = df1[df1['acquirer'].isin(list1)]
list2 = df3['acquirer'].tolist()
```

```python
df6 = df2[df2['CAPS'].isin(list2)]
df6.head()
```

```python
df3.head()
```

```python
df3 = df3.loc[df3['mr_info_full_name'] == 'Acquisition of shares']
df3.shape
```

```python
from datetime import datetime, timedelta
d = {'DATE OF ACQUISTION':[],'DATE one month AGO':[],'DATE ON YEAR ago':[]}
df4 = pd.DataFrame(data = d)
for index, row in df3.iterrows():
    currentdatetime = row['mgtkas_date']
    N = 365
    date_N_days_ago = currentdatetime - timedelta(days= 10)
    date_N_days_later = date_N_days_ago - timedelta(days=N)
    df4 = df4.append({'DATE OF ACQUISTION':currentdatetime,'DATE one month AGO':date_N_days_ago,'DATE ON YEAR ago':date_N_days_later}, ignore_index=True)
```


480 ACQUISTION CASES WE EXPECT 480 CSV FILES TO BE OUTPUT 

```python
df4.shape
```

```python
df5 = pd.merge(df3, df4, how='inner',  left_on='mgtkas_date', right_on = 'DATE OF ACQUISTION')
df5.shape
```

```python
writer = pd.ExcelWriter('result.xlsx')
df5.to_excel(writer,'Sheet1')
writer.save()
```
