```python
import pandas as pd
data = pd.read_excel('result.xlsx')
list = data['Scrip Code'].tolist()
len(list)
```

```python
list2 = []
i = 0
while(i<480):
    current = list[i]
    count = 0
    new = ''
    if(i>=1):
     j = 0
     while(j<i):
            if(current == list[j]):
                count = count + 1
            j = j + 1
     if(count==0):
        new = str(current) + '.csv'
     else:
        new = str(current)+' '+'(' + str(count)+ ')' + '.csv'
        
    if(i==0):
        new = str(current)+'.csv'
    list3 = []
    list3.append(new)
    list2.append(list3)
    i = i + 1
```

```python
len(list2)
import csv
with open('filename.csv','a') as csvFile:
    writer = csv.writer(csvFile)
    i = 0
    while(i<480):
        writer.writerow(list2[i])
        i = i + 1

```
